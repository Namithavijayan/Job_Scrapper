```python
import pandas as pd
import time
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
from selenium.common.exceptions import NoSuchElementException
import re
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support import expected_conditions as expect
import pandas as pd
from IPython.display import HTML
```

```python

driver = webdriver.Chrome( ChromeDriverManager().install())
```

```python
driver.get('https://www.naukri.com/ml-engineer-jobs?k=ml%20engineer&nignbevent_src=jobsearchDeskGNB')
```

```python
No_job_cards =10
```

```python
username = "namithavijayan24@gmail.com"
password = "Ammu@2001"
time.sleep(3)
driver.find_element(By.ID, "login_Layer").click()
print("h")

WebDriverWait(driver, 5).until(
        expect.visibility_of_element_located(
        (By.XPATH,"//input[@placeholder='Enter your active Email ID / Username']")))
driver.find_element(By.XPATH,"//input[@placeholder='Enter your active Email ID / Username']").send_keys(username)
driver.find_element(By.XPATH,"//input[@placeholder='Enter your password']").send_keys(password)
driver.find_element(By.CSS_SELECTOR, ".naukri-drawer .drawer-wrapper>.login-layer form .loginButton").click()
```

```python


job_list = []

count_job = 0
job_cards =  driver.find_elements(By.CSS_SELECTOR, '.srp-jobtuple-wrapper .sjw__tuple ')
for job_card in job_cards[:No_job_cards]:

        job =[]
        posted_at = None
        title = None
        experience = None
        location = None 
        apply_link = None
        #Reading title from main webpage  
        try:
            title_element = job_card.find_element(By.CSS_SELECTOR, '.row1')
            title = title_element.text
        except NoSuchElementException:
            pass 
        # Reading posted at details
        try:
           posted_at_element = job_card.find_element(By.CSS_SELECTOR, ".job-post-day")
           posted_at = posted_at_element.text
        except NoSuchElementException:
            pass

        # Reading experience required
        try:
            experience_element = job_card.find_element(By.CSS_SELECTOR, '.expwdth')
            experience = experience_element.text
        except NoSuchElementException:
            pass 
        #Reading location
        try:
            loc_element = job_card.find_element(By.CSS_SELECTOR, '.locWdth')
            location = loc_element.text
        except NoSuchElementException:
            pass 
        #Switching to new tab for extracting details
        URL = job_card.find_element_by_tag_name('a').get_attribute('href')
        driver.execute_script("window.open('');")
        driver.switch_to.window(driver.window_handles[1])
        driver.get(URL)
        time.sleep(3)
        try:
            
            job_description =  driver.find_element(By.CSS_SELECTOR, '.styles_job-desc-container__txpYf .styles_JDC__dang-inner-html__h0K4t')
            description = job_description.text
        except NoSuchElementException:
            pass
        
        # Accessing apply link in new tab
        try:
             driver.find_element(By.ID,'company-site-button').click()
             driver.switch_to.window(driver.window_handles[2])
             apply_link = driver.current_url
             driver.close()
             driver.switch_to.window(driver.window_handles[1])
        except NoSuchElementException:
            apply_link = driver.current_url
            pass      

        driver.close()
        driver.switch_to.window(driver.window_handles[0])
        

        job.append( posted_at)
        job.append(title)
        job.append(experience)
        job.append(location)
        job.append(description)
        job.append(apply_link)
        job_list.append(job)
    
    
          


print(job_list)
```

```python
pd.set_option('display.max_colwidth', None)
df = pd.DataFrame(job_list, 
                  columns = ['posted_at', 'title', 'experience' , 'location','description','apply_link'])

df = df.replace('\n',' ', regex=True)
HTML(df.to_html(render_links=True, escape=False))
```
